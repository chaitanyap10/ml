```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('A5-Food-Truck.csv')
data.plot(x="xlabel",y="ylabel",style="o")
plt.show()

x_mean=data["xlabel"].mean()
y_mean=data["ylabel"].mean()

data["x"]=data["xlabel"]-x_mean
data["y"]=data["ylabel"]-y_mean
data["x*y"]=data["x"]*data["y"]
data["x^2"]=data["x"]**2
data["y^2"]=data["y"]**2

sum_x_y=data["x*y"].sum()
sum_x_squared=data["x^2"].sum()
sum_y_squared=data["y^2"].sum()

correlation=sum_x_y/(sum_x_squared*sum_y_squared)
print(correlation)


def getMean(numbers):
    if len(numbers) == 0:
        return None
    else:
        current_sum = 0
        for i in numbers:
            current_sum += i
            current_avg = current_sum / len(numbers)
        return current_avg
    
def getStandardDeviation(numbers):
    if len(numbers) == 0:
        return 0
    else:
        mean = getMean(numbers)
        std_deviation = 0
        for i in numbers:
            std_deviation += (i - mean) ** 2
        return (std_deviation / len(numbers)) ** 0.5
    
std_deviation_x = getStandardDeviation(data["x"].tolist())
std_deviation_y = getStandardDeviation(data["y"].tolist())
print(std_deviation_x, std_deviation_y)

m= correlation*(std_deviation_y/std_deviation_x)

c=data["ylabel"].mean()-m*data["xlabel"].mean()


data["y_pred"]=m*data["xlabel"]+c


plot1 = plt.scatter(data["xlabel"], data["ylabel"])
plot2 = plt.scatter(data["xlabel"], data["y_pred"])
print(plt.show())

data["costvals"]=(data["y_pred"]-data["ylabel"])**2
cost=(data["costvals"].sum())/10
print("cost",cost)

sse=data["costvals"].sum()
print("sse",sse)

data["ssrvals"] = (data["y_pred"] - y_mean) ** 2

ssr = data["ssrvals"].sum()
print("ssr",ssr)

sst = ssr + sse
print("sst",sst)
rSquared = 1 - (sse / sst)
print("r square",rSquared)




```
